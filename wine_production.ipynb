```python
import cloudpickle
import numpy as np
import pandas as pd
```

```python
pipe = cloudpickle.load(open('wine.pkl', 'rb'))
```

```python
def prediction(arr):
    cols = ['alcohol', 'malic_acid', 'ash', 'alcalinity_of_ash', 'magnesium',
       'total_phenols', 'flavanoids', 'nonflavanoid_phenols',
       'proanthocyanins', 'color_intensity', 'hue',
       'od280/od315_of_diluted_wines', 'proline']
    
    df = pd.DataFrame([arr], columns=cols)
    return pipe.predict(df)[0]
```

```python
features = prediction([13.51,	1.80,	2.65,	19.0,	110.0,	2.35,	2.53,	0.29,	1.54,	4.20,	1.10,	2.87,	1095.0])
print(f"Target: {features:.2f}")
```
